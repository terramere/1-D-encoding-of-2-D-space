```python
def intersperse(first,second):
    min_len = min(len(first), len(second))
    return "".join([first[i] + second[i] for i in range(min_len)]) + first[min_len:] + second[min_len:]
```

```python
def morton1(mort_grid,d):
  # set morton numbers in grid
  for i in range(0,d):
    for j in range(0,d):
      b_row = bin(i)[2:].zfill(d)
      b_col = bin(j)[2:].zfill(d)
      bin_cell = intersperse(b_row,b_col)
      #bin_cell = intersperse(bin(j)[2:],bin(i)[2:])
      mort_grid[i,j] = int(bin_cell, 2)
  return mort_grid
```
